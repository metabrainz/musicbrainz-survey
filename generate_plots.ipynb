```python
import pandas as pd
#import numpy as np
import colorlover as cl

from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

```python
general = pd.read_csv("mb_usersurvey_2017_general.csv")
```

```python
years_used = general["For how many years have you been using it?"]
iplot([go.Histogram(y=years_used)], filename='pandas/simple-histogram')
```

```python
getting_started = general["How easy was it for you to get started adding data to MB?"]
getting_started_sums = pd.value_counts(getting_started.values).sort_index()

colour_scale_r2g = cl.scales["5"]["div"]["RdYlGn"]

# This probably would've been easier using a histogram but where's the fun in that?
data=go.Bar(
    y=getting_started_sums,
    marker=dict(
        color=color_scale_r2g
    )
)

layout = go.Layout(
    title="How easy was it for you to get started adding data to MB?",
    yaxis=dict(
        title="Count"
    ),
    bargap=0.1,
    bargroupgap=0.1
)

fig = go.Figure(data=[data], layout=layout)

iplot(fig, filename='styled barchart')
```

```python
current_state = general["How happy are you with MusicBrainz in its current state?"]

data=go.Histogram(
    x=current_state,
    marker=dict(
        color=color_scale_r2g
    )
)

layout = go.Layout(
    title="How happy are you with MusicBrainz in its current state?",
    yaxis=dict(
        title="Count"
    ),
    bargap=0.1,
    bargroupgap=0.1
)

fig = go.Figure(data=[data], layout=layout)

iplot(fig, filename='styled histogram')
```

```python
languages = general["What interface language do you use MB in?"]
languages_sums = pd.value_counts(languages.values)


trace = go.Pie(labels=list(languages_sums.index), values=languages_sums.values)
iplot([trace], filename='basic_pie_chart')
```
