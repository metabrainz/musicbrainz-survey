```python
import pandas as pd
#import numpy as np
import colorlover as cl

from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

```python
general = pd.read_csv("mb_usersurvey_2017_general.csv")
```

```python
years_used = general["For how many years have you been using it?"]
iplot([go.Histogram(y=years_used)], filename='pandas/simple-histogram')
```

```python
users_per_year = pd.value_counts(years_used.values).sort_index(ascending=False)
cumultative_users = users_per_year.cumsum()
cumultative_users_avg = cumultative_users.rolling(window=3, center=False).mean()

p1 = go.Scatter(
    name="Cumultative Users",
    x=list(cumultative_users.index),
    y=cumultative_users,
    fill="tonexty",
    marker = dict(
        color="#C51B8A",
        #opacity="50%"
    )
)

p2 = go.Scatter(
    name="Rolling average",
    x=list(cumultative_users_avg.index),
    y=cumultative_users_avg,
    fill="tonexty",
    marker = dict(
        opacity="50%"
    )
)

layout = go.Layout(
    xaxis=dict(
        range=[17, 0]
    )
)

fig = go.Figure(data=[p1], layout=layout)

iplot(fig)
```

```python
getting_started = general["How easy was it for you to get started adding data to MB?"]
#getting_started_sums = pd.value_counts(getting_started.values).sort_index()

color_scale_r2g = cl.scales["5"]["div"]["RdYlGn"]

plot = go.Histogram(
    x = getting_started,
    marker = dict(
        color = color_scale_r2g
    )
)

layout = go.Layout(
    title="How easy was it for you to get started adding data to MB?",
    yaxis=dict(
        title="Count"
    ),
    bargap=0.1,
    bargroupgap=0.1
)

fig = go.Figure(data=[plot], layout=layout)

iplot(fig)
```
